```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# change the datatype of release date to datetime in year

df['Release_Date'] = pd.to_datetime(df['Release_Date'])
print(df['Release_Date'].dtypes)
```

```python
df['Release_Date'] = df['Release_Date'].dt.year
df['Release_Date'].dtypes
```

```python
df.Release_Date.head()
```

```python
# drop the columns

cols = ['Overview','Original_Language','Poster_Url']
df.drop(cols,axis=1,inplace=True)
df.columns
```

```python
df.head()
```

categorizing **vote_average** column into 4 categories: **popular** , **average** , **below_avg**. **not_popular**

```python
def catigorize_col(df,col,labes):
  edges = [df[col].describe()['min'],
           df[col].describe()['25%'],
           df[col].describe()['50%'],
           df[col].describe()['75%'],
           df[col].describe()['max']]
  df[col] = pd.cut(df[col],edges,labels=labes,duplicates='drop')
  return df
```

```python
labels = ['not_popular','below_avg','average','popular']

catigorize_col(df,'Vote_Average',labels)

df['Vote_Average'].unique()
```

```python
df.head()
```

```python
df['Vote_Average'].value_counts()
```

```python
df.dropna(inplace=True)

df.isna().sum()
```

## split genres into a list

```python
df['Genre'] = df['Genre'].str.split(',')

df = df.explode('Genre').reset_index(drop=True)

df.head()
```

```python
# casting column into cateogy

df['Genre'] = df['Genre'].astype('category')
df['Genre'].dtypes
```

# **Data Visualization**

# Q1. What is the most frequent genre of movies released on Netflix?

```python
df['Genre'].describe()
```

```python
sns.catplot(y = 'Genre', data = df, kind = 'count',
            order = df['Genre'].value_counts().index,
            color = 'blue')
plt.title("Genre column distrubution")
plt.show()
```

# Q2. What genres has highest votes?

```python
df.head()
```

```python
sns.catplot(y = 'Vote_Average', data = df, kind = 'count',
            order = df['Vote_Average'].value_counts().index,
            color = 'blue')

plt.title('votes distribution')

plt.show()
```
